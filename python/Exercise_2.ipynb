```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Homo_sapiens.GRCh38.107.gtf.gz',
                 sep ="\t", 
                 header = None, 
                 comment='#', 
                 compression='gzip')
df.head()
```

```python
df.columns = ['seqname', 'source', 'feature', 'start', 'end', 'score', 'strand', 'frame', 'group']
```

```python
df
```

```python
df["source"].value_counts()
```

```python
ens_df = df[df["df"].str.contains("ensembl_havana")].head()
```

```python
ens_df = df.query("source == 'ensembl_havana'")
```

```python
ens_df.info()
```
